```python
import pandas as pd
```

```python

df = pd.read_csv("Sp20-OB-581-01_02.csv")
```

```python
name_dict = {}

for row, data in df.iterrows():
    name = data["Name"].split(", ")
    first_stem = name[1].split(" ")
    #print(len(first_stem), first_stem)
    if len(first_stem)>1:
        first = first_stem[0]
    else:
        first = name[1]
    name_dict[data["SUNet ID"]] = first + " "+ name[0][0]
```
